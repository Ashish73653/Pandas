## Arithmetic operation

```python
import pandas as pd

var = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var
```

```python
var["C"] = var["A"]+var["B"]
var
```

```python
var["C"] = var["A"]-var["B"]
var
```

```python
var["C"] = var["A"]*var["B"]
var
```

```python
var["C"] = var["A"]/var["B"]
var
```

```python
var["C"] = var["A"]%var["B"]
var
```

```python
var["C"] = var["A"]**var["B"]
var
```

```python
var1 = pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
var1
```

```python
var1["less than 25"] = var1["A"]<=25
var1
```

```python
var1["greater than 25"] = var1["B"]>=25
var1
```
